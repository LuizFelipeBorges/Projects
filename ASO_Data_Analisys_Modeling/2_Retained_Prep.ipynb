```
import pandas as pd
import numpy as np
```

```
retained = pd.read_excel('/content/Retained_Bank.xlsx', parse_dates= ['Date'])
retained
```

 </br>

## Limpeza e Tratamento
- Os dados que não correspondiam aos canais organicos, e a coluna *Acquisition Channel* foi retirada, pois se tornaram redundantes.
- A data foi formatada

```
retained_organic = retained[retained['Acquisition Channel'] == 'Organic']
retained_organic = retained_organic.drop(columns= ['Acquisition Channel'])
retained_organic['Date'] = retained_organic['Date'].map(lambda x: x.replace(" 00:00:00", ""))
retained_organic.index = range(retained_organic.shape[0])
retained_organic
```

## Exportação da Base
Nesse caso será exportada somente em CSV, pois são poucos os tratamentos necessários.

```
retained_organic.to_csv('retained_bank_organic.csv', index= False)
```
