```
import pandas as pd
import numpy as np

import requests as rq
import bs4 as bs4
import json
from google.colab import drive
drive.mount('/content/drive')
```

```
buscas = {"12": "Etios+Hatch", "19": "HB20", "5": "Fiesta+Hatch", "5": "Ka", 
          "15": "C3", "3": "Up!", "3": "Fox", "3": "Polo+Hatch"}

url = "https://www.shopcar.com.br/busca.php?tipo=1&marca={}&string={}&ordenar=valor_desc&pagina={}"
```

# Coleta dos dados

12 Toyota <br>
19 Hyundai <br>
5 Ford <br>
15 Citroen


```
for key, value in buscas.items():
  for page in range(1,5):
    urll = url.format(key, value, page)
    response = rq.get(urll) 

    with open("/content/drive/My Drive/Projetos/Rec_carros/Scraper_busca/{}_{}.html".format(value, page), 'w+') as output:
      output.write(response.text)
```

# Processamento dos dados brutos

```
for key, value in buscas.items():
  for page in range(1,5):
    with open("/content/drive/My Drive/Projetos/Rec_carros/Scraper_busca/{}_{}.html".format(value, page), 'r+') as inp:
      page_html = inp.read()

      parsed = bs4.BeautifulSoup(page_html)
            
      tags = parsed.findAll("a", "link")

      for e in tags:
          link = e['href']

          with open("/content/drive/My Drive/Projetos/Rec_carros/Scraper_busca/Links_carros/parsed_links.json", 'a+') as output:
            data = {"link": link}
            output.write("{}\n".format(json.dumps(data)))
  
```
