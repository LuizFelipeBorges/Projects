# Tratando os dados

## Tratando a coluna Modelo

```
df['modelo'] = df['modelo'].map(lambda x: x.replace("-", " "))
df.set_index('id', inplace= True)
df.head()
```

## Tratando a coluna Valor

```
def remove_sifrao(valor):
  v = valor.replace("R$", "")
  return v

def tira_virgula(valor):
  v = valor.replace(",", "")
  return v
```

```
df["valor"] = df["valor"].map(remove_sifrao)
df["valor"] = df["valor"].map(tira_virgula)
df["valor"] = pd.to_numeric(df["valor"])
df["valor"] = df["valor"]*1000
df
```

## Tratando a coluna KM

```
df["km"] = df["km"]*1000
df
```

```
df.describe()
```

#Modelo Base

```
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, average_precision_score
```

```
colunas = ['valor','km']

X= df[colunas]
y= df['y']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.5, random_state = 0)

mdl = DecisionTreeClassifier(random_state=0, max_depth=2, class_weight="balanced")
mdl.fit(X_train, y_train)

p = mdl.predict_proba(X_valid)[:, 1]

print(average_precision_score(y_valid, p))
print(roc_auc_score(y_valid, p))
```
